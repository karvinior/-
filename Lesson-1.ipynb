```python
import requests
import json
from pprint import pprint
```

### 1. Посмотреть документацию к API GitHub, разобраться как вывести список репозиториев для конкретного пользователя, сохранить JSON-вывод в файле *.json.

```python
main_link = 'https://api.github.com'    # выбираем, что будем парсить
user = 'karvinior'    # выбираем юзера. В примере - это я, на там ничего интеерсно :)
```

```python
response = requests.get(f'{main_link}/users/{user}/repos')    # скачиваем инфо c GitHub о репозиториях пользователе
```

```python
if response.ok:    # если успешно сохранено
    data = json.loads(response.text)    # сохранение в json (тип словарь)
```

```python
pprint(data)    # пришлось руками искать под каким ключом сохранены имена репощиториев
```

```python
for _ in data:
    print(_['name'])    # вывод имени репозитория
```
